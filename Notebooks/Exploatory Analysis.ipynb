```python
#6.2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: set plotting style
sns.set(style="whitegrid")
# Load cleaned and slimmed dataset
slim_df = pd.read_csv('clean data/airbnb_listings_slim.csv', parse_dates=['date'])
print(slim_df.shape)
slim_df.head()
```

```python
#Select Variables for EDA
numeric_cols = ['price', 'accommodates', 'bedrooms', 'beds', 'bathrooms',
                'minimum_nights', 'maximum_nights', 'number_of_reviews',
                'review_scores_rating', 'review_scores_cleanliness',
                'review_scores_location', 'reviews_per_month',
                'host_listings_count', 'availability_365']

categorical_cols = ['room_type', 'neighbourhood_cleansed', 
                    'instant_bookable', 'cancellation_policy', 'host_is_superhost']
```

```python
# 1ï¸âƒ£ Price distribution
plt.figure(figsize=(10,6))
sns.histplot(slim_df['price'], bins=50)
plt.xlim(0, 1000)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()  # show first
plt.savefig("price_distribution.png", dpi=300, bbox_inches='tight')
plt.close()

```

```python
# 1 = perfect positive correlation

#-1 = perfect negative correlation

#0 = no linear correlation
#The histogram of price shows that most listings are concentrated under â‚¬1000, with a long tail of very expensive properties.
#Insight: There are some extreme outliers in price that may skew averages.
#Action: For analysis, we may limit plots to a max of â‚¬1000 to focus on the main cluster.
```

```python
# 2ï¸âƒ£ Average price by room type
plt.figure(figsize=(8,6))
sns.barplot(x='room_type', y='price', data=slim_df)
plt.title("Average Price by Room Type")
plt.ylabel("Price")
plt.xlabel("Room Type")
plt.show()  # show first
plt.savefig("avg_price_by_room_type.png", dpi=300, bbox_inches='tight')
plt.close()

#Entire home/apt is the most expensive on average.
#Private room is moderately priced.
#Shared room is the least expensive.
#Insight: Room type strongly affects price, which is expected.
```

```python
# 3ï¸âƒ£ Correlation matrix heatmap
numeric_cols = [
    'price', 'accommodates', 'bedrooms', 'beds', 'bathrooms',
    'minimum_nights', 'maximum_nights', 'number_of_reviews',
    'review_scores_rating', 'review_scores_cleanliness',
    'review_scores_location', 'reviews_per_month', 
    'host_listings_count', 'availability_365'
]

# Compute correlations
corr = slim_df[numeric_cols].corr()

# Plot
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")

# âœ… Save BEFORE showing
plt.savefig(
    "/Users/am/Desktop/1_CF_Data_Analysis_Essentials/2_Data Analytics_Immersions_2 von_3/6/6.1/6_Airbnb/02_Data/Clean/correlation_matrix.png",
    dpi=300,
    bbox_inches='tight'
)

# âœ… Then show
plt.show()

# Close to free memory
plt.close()
#The heatmap shows relationships between numeric variables. Key correlations:
#price vs accommodates and price vs bedrooms â†’ moderately positive correlations.
#review_scores_rating vs review_scores_cleanliness â†’ strongly positive, as expected.
#Other variables show weak or negligible correlations.
#Interpretation: Positive correlations indicate that as one variable increases, the other tends to increase as well. Negative correlations indicate an inverse relationship.
```

```python
# 4ï¸âƒ£ Scatterplots for strong correlations
plt.figure(figsize=(8,6))
sns.scatterplot(x='accommodates', y='price', data=slim_df, alpha=0.3)
plt.title("Price vs Accommodates")
plt.savefig("scatter_price_accommodates.png", dpi=300, bbox_inches='tight')
plt.close()

plt.figure(figsize=(8,6))
sns.scatterplot(x='bedrooms', y='price', data=slim_df, alpha=0.3)
plt.title("Price vs Bedrooms")
plt.show()  # show first
plt.savefig("scatter_price_bedrooms.png", dpi=300, bbox_inches='tight')
plt.close()

#Scatterplots of:
#Price vs Accommodates
#Price vs Bedrooms
#Show that larger apartments (more accommodates or bedrooms) tend to have higher prices.
#Observation: There is some spread, suggesting other factors (like location or amenities) also influence price.
```

```python
# 5ï¸âƒ£ Categorical plot (boxplot) by room type

# Create the plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='room_type', y='price', data=slim_df)
plt.ylim(0, 1000)
plt.title("Price Distribution by Room Type")

# âœ… Save before showing
plt.savefig(
    "/Users/am/Desktop/1_CF_Data_Analysis_Essentials/2_Data Analytics_Immersions_2 von_3/6/6.1/6_Airbnb/02_Data/Clean/boxplot_price_room_type.png",
    dpi=300,
    bbox_inches='tight'
)

# âœ… Then show it in the notebook
plt.show()

# Optional: close after saving
plt.close()
#The boxplot of price by room_type shows:
#High variation within room types.
#Entire homes have the widest price range.
#Shared rooms rarely exceed â‚¬200.
#Conclusion: Room type is a strong determinant of price, but there are outliers and variability within categories
```
